```python
## Left/right join
```

```python
movies = pd.read_pickle('movies.p')
```

```python
movies_to_genres = pd.read_pickle('movie_to_genres.p')
```

```python
movies.columns
```

```python
movies_to_genres.columns
```

```python
movies.head()
```

```python
movies_to_genres.head()
```

```python
movies.merge(movies_to_genres, left_on='id', right_on='movie_id', how='left')
```

```python
movies.merge(movies_to_genres, left_on='id', right_on='movie_id', how='right')
```

```python
movies.merge(movies_to_genres, left_on='id', right_on='movie_id', how='outer')
```
