```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'temperatures.txt')
```

```python
data = data.drop(['Unnamed: 0'],axis=1)
data
```

```python
data.drop_duplicates(subset='city', keep='first')
```

```python
data.drop_duplicates(subset='country', keep='first')
```

```python
data.groupby('country').first()
```

```python
data.groupby('country')['avg_temp_c'].max()
```

```python
plt.figure(figsize=(24,6))
plt.xticks(rotation = 45)
plt.plot(data.groupby('country')['avg_temp_c'].max())
```

```python
data.groupby('country')['avg_temp_c'].min()
```

```python
plt.figure(figsize=(24,6))
plt.xticks(rotation = 45)
plt.plot(data.groupby('country')['avg_temp_c'].min())
```

```python
indiaData = data[data['country']=='India']
indiaData
```

```python
plt.figure(figsize=(24,6))
plt.xticks(rotation = 45)
plt.plot(indiaData.groupby('city')['avg_temp_c'].min())
```
